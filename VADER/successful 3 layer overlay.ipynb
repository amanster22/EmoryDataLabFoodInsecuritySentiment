```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
privateorg_df = pd.read_csv('datasets/combined/privateorg_combined_output.csv')
government_df = pd.read_csv('datasets/combined/government_combined_output.csv')
nonprofit_df = pd.read_csv('datasets/combined/nonprofit_combined_output.csv')

# Add a source column to each dataset
privateorg_df['source'] = 'PrivateOrg'
government_df['source'] = 'Government'
nonprofit_df['source'] = 'NonProfit'

# Combine the datasets
combined_df = pd.concat([privateorg_df, government_df, nonprofit_df], ignore_index=True)

# Create the plot with only trend lines
plt.figure(figsize=(10, 6))
sns.kdeplot(data=combined_df, x='sentiment', hue='source', palette=['blue', 'magenta', 'orange'], fill=True, alpha=0.8)

# Add title and labels
plt.title('Count-based Distribution of Sentiment Scores (Trend Lines Only)')
plt.xlabel('Sentiment Score')
plt.ylabel('Density')

# Show the plot
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load datasets
privateorg_df = pd.read_csv('datasets/combined/privateorg_combined_output.csv')
government_df = pd.read_csv('datasets/combined/government_combined_output.csv')
nonprofit_df = pd.read_csv('datasets/combined/nonprofit_combined_output.csv')

# Add a source column to each dataset
privateorg_df['source'] = 'PrivateOrg'
government_df['source'] = 'Government'
nonprofit_df['source'] = 'NonProfit'

# Combine the datasets
combined_df = pd.concat([privateorg_df, government_df, nonprofit_df], ignore_index=True)

# Create the Plotly graph
fig = px.density_contour(
    combined_df,
    x='sentiment',
    color='source',
    marginal_x='histogram',  # Optional: Add marginal histograms
    marginal_y='violin',     # Optional: Add marginal violin plots
    template='plotly_white'
)

# Update layout for better visualization
fig.update_traces(contours_coloring="fill", opacity=0.3)
fig.update_layout(
    title='Proportional Distribution of Sentiment Scores (Interactive)',
    xaxis_title='Sentiment Score',
    yaxis_title='Density',
    legend_title='Source'
)

# Show the plot
fig.show()
```
