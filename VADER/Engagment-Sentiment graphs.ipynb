```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV file paths and corresponding labels
csv_files = [
    ('datasets/combined/government_combined_output.csv', 'Government'),
    ('datasets/combined/nonprofit_combined_output.csv', 'Nonprofit'),
    ('datasets/combined/privateorg_combined_output.csv', 'Private Organization')
]

# Columns to use for the y-axis
y_axes = ['LikesCount', 'SharesCount', 'CommentsCount']

# Loop through each CSV file
for csv_file, label in csv_files:
    # Load the data
    data = pd.read_csv(csv_file)
    
    # Ensure sentiment is within the range -1 to 1
    data = data[(data['sentiment'] >= -1) & (data['sentiment'] <= 1)]
    
    # Loop through each y-axis column
    for y_axis in y_axes:
        # Create a new figure for each graph
        plt.figure(figsize=(10, 6))
        
        # Scatter plot with sentiment on the x-axis and the current y-axis column
        plt.scatter(data['sentiment'], data[y_axis], alpha=0.6, color='blue')
        
        # Add labels and title
        plt.title(f'{label} - {y_axis} vs Sentiment', fontsize=14)
        plt.xlabel('Sentiment (Scale: -1 to 1)', fontsize=12)
        plt.ylabel(y_axis, fontsize=12)
        plt.grid(True)
        
        # Adjust layout and show the plot
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV file paths and corresponding labels
csv_files = [
    ('datasets/combined/government_combined_output.csv', 'Government'),
    ('datasets/combined/nonprofit_combined_output.csv', 'Nonprofit'),
    ('datasets/combined/privateorg_combined_output.csv', 'Private Organization')
]

# Columns to use for the y-axis
y_axes = ['LikesCount', 'SharesCount', 'CommentsCount']

# Colors for each y-axis column
colors = ['blue', 'green', 'red']

# Loop through each CSV file
for csv_file, label in csv_files:
    # Load the data
    data = pd.read_csv(csv_file)
    
    # Ensure sentiment is within the range -1 to 1
    data = data[(data['sentiment'] >= -1) & (data['sentiment'] <= 1)]
    
    # Create a figure with three subplots (stacked vertically)
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 18), sharex=True)
    
    # Loop through each y-axis column and corresponding color
    for i, (y_axis, color) in enumerate(zip(y_axes, colors)):
        # Scatter plot with sentiment on the x-axis and the current y-axis column
        axes[i].scatter(data['sentiment'], data[y_axis], alpha=0.6, color=color)
        
        # Set the y-axis to a logarithmic scale
        axes[i].set_yscale('log')
        
        # Add labels and title for each subplot
        axes[i].set_title(f'{label} - {y_axis} (Log Scale)', fontsize=12)
        axes[i].set_ylabel(f'{y_axis} (Log Scale)', fontsize=10)
        axes[i].grid(True, which='both', linestyle='--', linewidth=0.5)  # Add grid for both major and minor ticks
    
    # Add a shared x-axis label
    axes[-1].set_xlabel('Sentiment (Scale: -1 to 1)', fontsize=12)
    
    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV file paths and corresponding labels
csv_files = [
    ('datasets/combined/government_combined_output.csv', 'Government'),
    ('datasets/combined/nonprofit_combined_output.csv', 'Nonprofit'),
    ('datasets/combined/privateorg_combined_output.csv', 'Private Organization')
]

# Columns to use for the y-axis
y_axes = ['LikesCount', 'SharesCount', 'CommentsCount']

# Colors for each y-axis column
colors = ['blue', 'green', 'red']

# Loop through each CSV file
for csv_file, label in csv_files:
    # Load the data
    data = pd.read_csv(csv_file)
    
    # Ensure sentiment is within the range -1 to 1
    data = data[(data['sentiment'] >= -1) & (data['sentiment'] <= 1)]
    
    # Create a figure for the overlayed plot
    plt.figure(figsize=(10, 6))
    
    # Loop through each y-axis column and corresponding color
    for y_axis, color in zip(y_axes, colors):
        # Scatter plot with sentiment on the x-axis and the current y-axis column
        plt.scatter(data['sentiment'], data[y_axis], alpha=0.6, color=color, label=y_axis)
    
    # Set the y-axis to a logarithmic scale
    plt.yscale('log')
    
    # Add labels, title, and legend
    plt.title(f'{label} - Likes, Shares, and Comments vs Sentiment (Log Scale)', fontsize=14)
    plt.xlabel('Sentiment (Scale: -1 to 1)', fontsize=12)
    plt.ylabel('Counts (Log Scale)', fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Add grid for both major and minor ticks
    plt.legend()  # Add a legend to differentiate the columns
    
    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```
