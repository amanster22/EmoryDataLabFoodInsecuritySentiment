```python
import matplotlib.pyplot as plt

labels = ['Experience-Based', 'Opinion-Based']
positive = [20, 22]
neutral = [34, 39]
negative = [178, 87]
total = [sum(x) for x in zip(positive, neutral, negative)]

x = range(len(labels))
bar_width = 0.5

fig, ax = plt.subplots(figsize=(10, 6))

p1 = ax.bar(x, negative, bar_width, label='Negative', color='#d9534f')
p2 = ax.bar(x, neutral, bar_width, bottom=negative, label='Neutral', color='#f0ad4e')
p3 = ax.bar(x, positive, bar_width, bottom=[i+j for i,j in zip(negative, neutral)], label='Positive', color='#5cb85c')

# Labeling each bar segment with count and %
for idx in range(len(x)):
    ax.text(x[idx], negative[idx] / 2, f"{negative[idx]} ({negative[idx]/total[idx]*100:.1f}%)", ha='center', color='white', fontsize=10)
    ax.text(x[idx], negative[idx] + neutral[idx]/2, f"{neutral[idx]} ({neutral[idx]/total[idx]*100:.1f}%)", ha='center', color='black', fontsize=10)
    ax.text(x[idx], total[idx] - positive[idx]/2, f"{positive[idx]} ({positive[idx]/total[idx]*100:.1f}%)", ha='center', color='white', fontsize=10)

ax.set_ylabel('Number of Comments')
ax.set_title(' Government Sentiment Breakdown by Comment Type (380 Comments)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.set_ylim(0, max(total) + 40)

# Total count annotations
for idx, val in enumerate(total):
    ax.text(x[idx], total[idx] + 5, f"Total: {val}", ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['Government', 'Nonprofit', 'Private', 'Control']
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

experience = [8.6, 63.7, 31.6, 27.8]
opinion = [14.9, 56.7, 22.9, 26.9]
experience += experience[:1]
opinion += opinion[:1]

# Plot setup
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Plot lines and fill
ax.plot(angles, experience, color="#5cb85c", linewidth=2.5, label='Experience-Based Sentiment')
ax.fill(angles, experience, color="#5cb85c", alpha=0.25)

ax.plot(angles, opinion, color="#0275d8", linewidth=2.5, label='Opinion-Based Sentiment')
ax.fill(angles, opinion, color="#0275d8", alpha=0.25)

# Axis and grid
ax.set_thetagrids(np.degrees(angles[:-1]), labels, fontsize=12)
ax.set_ylim(0, 70)
ax.set_yticks([10, 20, 30, 40, 50, 60])
ax.set_yticklabels([f"{i}%" for i in [10, 20, 30, 40, 50, 60]], fontsize=10)
ax.set_rlabel_position(180)

# Add % labels near points (but cleaner)
for i in range(num_vars):
    ax.text(angles[i], experience[i]+3, f"{experience[i]:.1f}%", color='#5cb85c', fontsize=9, ha='center')
    ax.text(angles[i], opinion[i]-5, f"{opinion[i]:.1f}%", color='#0275d8', fontsize=9, ha='center')

# Legend
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), frameon=True)

# Clarification box (scale explanation)
plt.annotate(
    "Each ring = % of comments marked as positive sentiment.\nHigher values = stronger public trust/satisfaction.",
    xy=(0.5, -0.15), xycoords='axes fraction',
    ha='center', fontsize=9.5, bbox=dict(boxstyle="round,pad=0.5", fc="#f7f7f7", ec="gray")
)

# Narrative callouts  positioned outside the radar for clarity
plt.text(np.radians(15), 73, "Gov't has lowest\nexperience sentiment", color='red', fontsize=10, ha='left')
plt.text(np.radians(130), 73, "Nonprofits trusted\nmost across both types", color='green', fontsize=10, ha='right')

# Title
ax.set_title(" Positive Sentiment (%) by Sector & Comment Type", fontsize=15, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import ternary

# Set up the plot
scale = 100
fig, tax = ternary.figure(scale=scale)
fig.set_size_inches(8, 7)

# Define points
points = {
    'Government': (11.1, 19.2, 69.7),
    'Nonprofit': (60.6, 27.1, 12.3),
    'Private': (28.3, 34.8, 36.9),
    'Control': (27.3, 35.9, 36.8),
}

# Plot each point
for label, (pos, neu, neg) in points.items():
    tax.scatter([(neg, neu, pos)], marker='o', label=label, s=80)

# Customize the plot
tax.boundary(linewidth=2)
tax.gridlines(color="gray", multiple=10)
tax.left_axis_label("Neutral %")
tax.right_axis_label("Positive %")
tax.bottom_axis_label("Negative %")
tax.ticks(axis='lbr', linewidth=1, multiple=10, offset=0.02)
tax.legend(loc='upper right', fontsize=10)
tax.set_title("Sentiment Distribution Across Sectors", fontsize=14)

plt.tight_layout()
tax.show()
```
